```python
import json
import os
import cv2

dataset = {'categories': [], 'images': [], 'annotations': []}
# 根路径，里面包含images(图片文件夹)，annos.txt(bbox标注)，classes.txt(类别标签),以及annotations文件夹(如果没有则会自动创建，用于保存最后的json)
root_path = '/home/xsb/battery/±ÈÈüÎÄ¼þ/newtrain'
# 用于创建训练集或验证集 可以调成val
phase = 'battery_train'
# 训练集和验证集划分的界线  可以调
split = 1000
class_path = os.path.join(root_path, 'classes.txt')
print(class_path)
# 打开类别标签
with open(class_path) as f:
    classes = f.read().strip().split(',')

# 建立类别标签和数字id的对应关系 从零开始
for i, cls in enumerate(classes, 0):
    dataset['categories'].append({'id': i, 'name': cls, 'supercategory': 'mark'})

# 读取train2017文件夹的图片名称
# 改了，加了sorted，保证图片按名字升序，使得<'image_id': k >里面，k=0,对应第一张图片，00...01.jpg
file_path = os.path.join(root_path, 'images','train')
file_list = sorted(os.listdir(file_path))
_names = [f for f in file_list]
# 判断是建立训练集还是验证集
if phase == 'battery_train':
    names = [line for i, line in enumerate(_names) if i <= split]
elif phase == 'battery_val':
    names = [line for i, line in enumerate(_names) if i > split]

# 读取Bbox信息
with open(os.path.join(root_path, 'annos.csv')) as tr:
    annos = tr.readlines()

# 以上数据转换为COCO所需要的
for k, name in enumerate(names, 1):
    # 用opencv读取图片，得到图像的宽和高
    im = cv2.imread(os.path.join(file_path, name))
    height, width, _ = im.shape

    # 添加图像的信息到dataset中
    dataset['images'].append({'file_name': name,
                              'id': k,
                              'width': width,
                              'height': height})

    index = str(int(name[:-4]))
    # 一张图多个框时需要判断
    anno = annos[k-1]
    parts = anno.strip().split(',')
    # type(parts[0]) string
    # 如果图像的名称和标记的名称对上，则添加标记
    strlen = len(parts)
    if parts[0] == name:
        # 类别
        print(name)
        ii = 1
        for j in range(1, strlen, 5):
            if parts[j]== '':
                break
            else:
                cls_id = parts[j]
                # x_min
                x1 = float(parts[j+1])
                # y_min
                y1 = float(parts[j+2])
                x2 = float(parts[j+3])
                y2 = float(parts[j+4])
                width = x2 - x1
                height = y2 - y1
                dataset['annotations'].append({
                    'area': width * height,
                    'bbox': [x1, y1, width, height],
                    # 'category_id': int(cls_id),
                    'category_id': cls_id,
                    'id': ii,  # 一个图中标注的index
                    'image_id': k,
                    'iscrowd': 0,
                    # mask, 矩形是从左上角点按顺时针的四个顶点
                    'segmentation': [[x1, y1, x2, y1, x2, y2, x1, y2]]
                })
                ii = ii + 1



# 保存结果
# # folder = os.path.join(root_path, 'annotations')
folder = '/home/xsb/battery/±ÈÈüÎÄ¼þ/newtrain/annotations'

if not os.path.exists(folder):
    os.makedirs(folder)
json_name = os.path.join(folder, '{}.json'.format(phase))
with open(json_name, 'w') as f:
    json.dump(dataset, f)
print('done')
```
