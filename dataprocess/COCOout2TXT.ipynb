```python
import json
from collections import defaultdict

"""hyper parameters"""
json_file_path = '/home/xsb/文档/DFGX/test_bbox_results.json'
images_dir_path = '/home/r/xsb/dfgx/battery/images/test'
output_path = '/home/xsb/文档/DFGX/EffVal.txt'

"""load json file"""
'''
将coco格式的输出文件转换成比赛需要的标注格式
'''
name_box_id = defaultdict(list)
id_name = dict()
with open(json_file_path, encoding='utf-8') as f:
    data = json.load(f)
    print(data[0])
    pre_image_id = 0
    for item in data:
        image_id = item['image_id']
        print(image_id)
        category_id = item['category_id']
        a1,b1,a2,b2 = item['bbox']
        msg = "%d,%d,%d,%d,%d,"%(category_id,a1,b1,a2,b2)
        
        if not category_id:
            continue
            
        if image_id == pre_image_id:
            with open(output_path, 'a') as f:
                f.write(msg)
        else:
            with open(output_path, 'a') as f:
                f.write('\n')
                f.writelines([image_id,".jpg",","])
            pre_image_id = image_id
```
