```python
# 分割标签，第0个是文件名，因此从第1个开始每5个分割一次
def splitLabel(originalLabel):
    subLable = []
    subLable = [originalLabel[i:i + 5] for i in range(1, len(originalLabel), 5)]
    return subLable
```

```python
# 转换标签中的变换，原本给出的标签格式是
# 图片名，类别，左上x，左上y，左下x，左下y
# yolo需要的格式是
# 每个图片一个对应一个txt，内容是：
# 类别，归一化的：中心点x，中心点y，宽度w，高度y
def tranLabel(H, W, originalLabel):
    _class, a1, b1, a2, b2 = originalLabel
    _class, a1, b1, a2, b2 = int(_class), int(a1), int(b1),int(a2), int(b2)
    w = a2 - a1
    h = b2 - b1
#     x = (a2 + a1) / 2
#     y = (b2 + b1) / 2
    x = a1 + (a2 - a1) / 2
    y = b1 + (b2 - b1) / 2
    x = x/W
    y = y/H
    w = w/W
    h = h/H
    out = '%d %f %f %f %f'%(_class,x,y,w,h)
    return out
out = tranLabel(3504,5800,[1,5528,243,5558,334])
print(out)
```
